```python
import pandas as pd
```

```python
df = pd.read_csv('new_retail_data.csv')
```

```python
df.head()
```

```python
print("Shape of data:", df.shape)
print("\nColumn types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())
df.describe(include='all')
```

```python
df.dropna(subset=['Customer_ID', 'Transaction_ID'], inplace=True)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Hour'] = pd.to_datetime(df['Time'], errors='coerce').dt.hour
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Ratings'].fillna(df['Ratings'].median(), inplace=True)
df.shape
```

```python
print("Shape of data:", df.shape)
print("\nColumn types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())
df.describe(include='all')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
top_countries = df['Country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Countries by Orders")
plt.show()
```

```python
sns.histplot(df['Hour'].dropna(), bins=24, kde=False)
plt.title("Orders by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Orders")
plt.show()
```

```python
sns.scatterplot(x='Income', y='Total_Amount', hue='Gender', data=df)
plt.title("Income vs Total Purchase Amount")
plt.show()
```

```python
rfm = df.groupby('Customer_ID').agg({
    'Date': lambda x: (df['Date'].max() - x.max()).days,
    'Transaction_ID': 'nunique',
    'Total_Amount': 'sum'
})
rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm.dropna(inplace=True)

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)

kmeans = KMeans(n_clusters=4, random_state=42)
rfm['Segment'] = kmeans.fit_predict(rfm_scaled)

sns.scatterplot(data=rfm, x='Recency', y='Monetary', hue='Segment', palette='Set2')
plt.title("Customer Segments Based on RFM")
plt.show()
```

```python
category_perf = df.groupby('Product_Category').agg({
    'Ratings': 'mean',
    'Total_Amount': 'sum',
    'Transaction_ID': 'count'
}).sort_values(by='Total_Amount', ascending=False)

category_perf.columns = ['Avg_Rating', 'Total_Revenue', 'Order_Count']

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.scatterplot(data=category_perf, x='Order_Count', y='Total_Revenue', size='Avg_Rating', hue='Avg_Rating', sizes=(100, 1000), palette='coolwarm', legend=True)
plt.title("Product Category: Orders vs Revenue (Bubble = Avg Rating)")
plt.xlabel("Total Orders")
plt.ylabel("Total Revenue")
plt.show()
```
